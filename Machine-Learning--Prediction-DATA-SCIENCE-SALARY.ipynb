```python
#Gerekli kütüphaneleri import ettim
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
'''
work_year: Maaşın ödendiği yıl.
experience_level Aşağıdaki olası değerlerle yıl boyunca işteki deneyim seviyesi: 
    EN Giriş seviyesi / Junior MI Orta seviye / Orta SE Kıdemli seviye / Expert EX Yönetici seviyesi / Direktör
    
employment_type:PT Yarı zamanlı FT Tam zamanlı CT Sözleşmeli FL Serbest   

job_title: Yıl boyunca çalışılan rol.

salary:Ödenen toplam brüt maaş tutarı.

salary_currency:Bir ISO 4217 para birimi kodu olarak ödenen maaşın para birimi.

salaryinusd:USD cinsinden maaş

employee_residence:Çalışanın çalışma yılı boyunca ISO 3166 ülke kodu olarak birincil ikamet ettiği ülke.

remote_ratio:Uzaktan yapılan toplam iş miktarı, olası değerler aşağıdaki gibidir: 
    0 Uzaktan çalışma yok (%20'den az) 50 Kısmen uzak 100 Tamamen uzak (%80'den fazla)
    
company_location:ISO 3166 ülke kodu olarak işverenin ana ofisinin veya sözleşmeli şubesinin bulunduğu ülke.

company_size:Yıl içinde şirkette çalışan ortalama kişi sayısı: 
    S 50'den az çalışan (küçük) M 50 ila 250 çalışan (orta) L 250'den fazla çalışan (büyük)

'''
print("-")
```

```python
data=pd.read_csv("ds_salaries.csv")#veriyi hazırladım
data.head()
```

```python
data.drop("Unnamed: 0",axis=1,inplace=True)#"Unnamed: 0" sütununu veriden kaldırdım
#Label Encoder:
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
data["experience_level"]=le.fit_transform(data["experience_level"])
data["employment_type"]=le.fit_transform(data["employment_type"])
data["job_title"]=le.fit_transform(data["job_title"])
data["salary_currency"]=le.fit_transform(data["salary_currency"])
data["employee_residence"]=le.fit_transform(data["employee_residence"])
data["company_location"]=le.fit_transform(data["company_location"])
data["company_size"]=le.fit_transform(data["company_size"])
data.head()
```

```python
data.isnull().sum()#Veride boş değer olup olmadığını kontrol ettim
```

```python
x=data.drop("salary",axis=1)#Bağımsız değişken
y=data["salary"]#Bağımlı değişken

#Model hakkında bilgi edindim
model=sm.OLS(y,x).fit()
model.summary()
```

```python
#Veriyi test ve eğitim kümelerine ayırdım
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

#Standard Scaler işlemi Yaptım(Veriyi aynı dünyaya indirgemek)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lg=LinearRegression()
lg.fit(x_train,y_train)
y_pred=lg.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
#Polynomial Features
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=8)
x_poly=pf.fit_transform(x_train)
lg2=LinearRegression()
lg2.fit(x_poly,y_train)
y_pred=lg2.predict(pf.fit_transform(x_test))
print(r2_score(y_test,y_pred))
```

```python
#Support Vector Machine
from sklearn.svm import SVR
svr=SVR(kernel="rbf")
svr.fit(x_train,y_train)
y_pred=svr.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
#Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
dc=DecisionTreeRegressor()
dc.fit(x_train,y_train)
y_pred=dc.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
#Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=8)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
#XGBOOST
from xgboost import XGBRegressor
xgb=XGBRegressor()
xgb.fit(x_train,y_train)
y_pred=xgb.predict(x_test)
print(r2_score(y_test,y_pred))
```
